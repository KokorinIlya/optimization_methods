```python
import matplotlib.pyplot as plt
from HW1.condition_number import number_of_iters
from HW1.gradient_descent import linear_step_chooser, constant_step_chooser
from HW1.one_demensional import golden
import numpy as np
from tqdm.notebook import tqdm
```

```python
def make_plot(step_chooser):
    n_vars = [5, 10, 20, 50, 100]
    condition_numbers = np.linspace(1, 1000, 50)
    plt.figure()
    for var in tqdm(n_vars):
        iter_numbers = [number_of_iters(cond, var, step_chooser) for cond in condition_numbers]
        plt.plot(condition_numbers, iter_numbers, label='n={}'.format(var))

    plt.xlabel('Число обусловленности')
    plt.ylabel('Число итераций')
    plt.legend()
    plt.show()
```

```python
make_plot(linear_step_chooser(golden))
```

```python
make_plot(constant_step_chooser(1e-3))
```
